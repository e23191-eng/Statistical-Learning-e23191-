# **Gaussian Mixture Model Clustering as Conditional Updating**

**1. Deriving the Marginal Density**


Using the law of total probability, we can find the marginal density $p(x_i)$ by summing the joint probability density $p(x_i, C_i = k)$ over all possible latent cluster states $K$:
$$p(x_i) = \sum_{k=1}^K p(x_i, C_i = k) = \sum_{k=1}^K P(C_i = k) p(x_i \mid C_i = k)$$



Substituting the prior cluster membership probability $P(C_i=k)=\phi_k$ and the conditional multivariate Gaussian distribution $X_i\mid C_i=k \sim \mathcal{N}(\mu_k,\Sigma_k)$:
$$p(x_i)=\sum_{k=1}^K \phi_k \mathcal{N}(x_i\mid \mu_k,\Sigma_k)$$



**Interpretation:** This density is called a **Gaussian mixture density** because it is a weighted linear combination (or "mixture") of $K$ distinct Gaussian component densities , where the weights $\phi_k$ act as the mixing proportions.


**2. Deriving the Posterior Cluster Probability**


For a fixed observation $x_i$, Bayes' rule states:
$$P(C_i=k\mid X_i=x_i)=\frac{P(X_i=x_i\mid C_i=k)P(C_i=k)}{P(X_i=x_i)}$$



By applying the marginal density derived in Part 1 to the denominator, we get:
$$P(C_i=k\mid X_i=x_i)=\frac{ P(X_i=x_i\mid C_i=k)P(C_i=k) }{ \sum_{j=1}^K P(X_i=x_i\mid C_i=j)P(C_i=j) }$$



Substituting the known Gaussian components and priors:
$$P(C_i=k\mid X_i=x_i)=\frac{\phi_k\mathcal{N}(x_i\mid \mu_k,\Sigma_k) }{ \sum_{j=1}^K \phi_j\mathcal{N}(x_i\mid \mu_j,\Sigma_j) }$$



**Interpretation:** The responsibility $\gamma_{ik}$ is interpreted as a posterior probability because it represents our updated belief that data point $x_i$ belongs to cluster $k$ *after* evaluating the actual observed value $x_i$ against all component distributions[cite: 167, 196].


**3. One-Hot Encoding of the Latent Cluster Variable**


Given that $Z_{ik}$ is a binary indicator variable ($Z_{ik}=1$ if $C_i=k$ and $0$ otherwise):


$$\mathbb{E}[Z_{ik}\mid X_i=x_i] = (1 \cdot P(Z_{ik}=1 \mid X_i=x_i)) + (0 \cdot P(Z_{ik}=0 \mid X_i=x_i))$$
$$\mathbb{E}[Z_{ik}\mid X_i=x_i] = P(C_i=k \mid X_i=x_i) = \gamma_{ik}$$



Stacking these expectations into the vector $\mathbb{E}[Z_i\mid X_i=x_i]$ directly yields[cite: 169]:
$$\mathbb{E}[Z_i\mid X_i=x_i]=\begin{bmatrix} \gamma_{i1} \\ \gamma_{i2} \\ \vdots \\ \gamma_{iK} \end{bmatrix}$$



**Conclusion:** The soft cluster assignment vector in a Gaussian mixture model is precisely the conditional expectation $\mathbb{E}[Z_i\mid X_i=x_i]$ of the one-hot encoded cluster membership vector.



**4. From Soft Assignment to Hard Clustering**


* **Soft Clustering:** Gives a soft assignment of $x_i$ to all clusters via a fractional probability vector $\mathbb{E}[Z_i\mid X_i=x_i]$. The point belongs partially to multiple clusters simultaneously, capturing boundary ambiguity.
* **Hard Clustering:** Places the data point deterministically into a single cluster by choosing the cluster with the largest posterior probability: $\widehat{C}_i = \operatorname{arg\,max}_{1\le k\le K} \gamma_{ik}$ , stripping away uncertainty.



**5. Conditional Expectation Comparisons**


Given $X_i \mid C_i=k \sim \mathcal{N}(\mu_k, \Sigma_k)$, the mean of this Gaussian is explicitly $\mu_k$, so $\mathbb{E}[X_i\mid C_i=k]=\mu_k$.
* **$\mu_k$** can be interpreted as the geometric center of cluster $k$ in the data feature space.
* Comparing $\mathbb{E}[Z_i\mid X_i=x_i]$ and $\mathbb{E}[X_i\mid C_i=k]$, the first gives the soft cluster membership of an observed point, while the second gives the mean location of a cluster.



**6. The Complete-Data Likelihood**


The complete-data likelihood is defined as:


$$p(x_1,\dots,x_n,z_1,\dots,z_n)=\prod_{i=1}^n \prod_{k=1}^K \left[ \phi_k \mathcal{N}(x_i\mid \mu_k,\Sigma_k) \right]^{z_{ik}}$$

Taking the natural logarithm ($\log$) changes the products into sums, showing that the complete-data log-likelihood is:
$$\ell_c=\sum_{i=1}^n \sum_{k=1}^K z_{ik} \left[ \log \phi_k + \log \mathcal{N}(x_i\mid \mu_k,\Sigma_k) \right]$$

**Explanation:** This expression would be easy to maximize if the values of $z_{ik}$ were known because it decouples for each cluster component. We could perform standard, closed-form Maximum Likelihood Estimation (MLE) independently for each cluster (e.g., simple empirical sample means and variances for the points belonging to group $k$).



**7. The EM Interpretation**


In practice, the latent variables $Z_i$ are not observed. The EM algorithm replaces the unknown indicators $z_{ik}$ by their conditional expectations given the observed data and current parameter estimates: $z_{ik} \leadsto \mathbb{E}[Z_{ik}\mid X_i=x_i] = \gamma_{ik}$. This is the E-step of the EM algorithm, resulting in the expected complete-data log-likelihood:
$$Q=\sum_{i=1}^n \sum_{k=1}^K \gamma_{ik} \left[ \log \phi_k + \log \mathcal{N}(x_i\mid \mu_k,\Sigma_k) \right]$$



**Explanation:** The E-step can be interpreted as a **conditional update** of cluster membership probabilities  because it uses Bayes' theorem to recalculate the latent responsibilities ($\gamma_{ik}$) based on the newly updated parameter values.



**8. Parameter Updates**


By maximizing $Q$ with respect to the model parameters, we derive the standard GMM updates:


$$N_k=\sum_{i=1}^n \gamma_{ik}, \quad \phi_k^{\text{new}}=\frac{N_k}{n}, \quad \mu_k^{\text{new}}=\frac{1}{N_k} \sum_{i=1}^n \gamma_{ik}x_i$$
$$\Sigma_k^{\text{new}}=\frac{1}{N_k} \sum_{i=1}^n \gamma_{ik} (x_i-\mu_k^{\text{new}}) (x_i-\mu_k^{\text{new}})^T$$

**Explanation:** The responsibility $\gamma_{ik}$ acts as a **fractional membership weight** of observation $x_i$ in cluster $k$ [cite: 191] because instead of a data point contributing $1$ or $0$ (like in hard K-Means), its structural contribution to updating a cluster's mean and covariance matrix is scaled continuously by the model's posterior confidence ($\gamma_{ik}$).


**9. General Summary Interpretation**


Gaussian mixture clustering can be viewed as a repeated process of conditional updating for the following reasons:


* The mixture weight $\phi_k$ is the prior probability of cluster $k$.


* The Gaussian density $\mathcal{N}(x_i\mid \mu_k,\Sigma_k)$ measures how compatible $x_i$ is with cluster $k$.


* The responsibility $\gamma_{ik}$ is the posterior probability of cluster $k$ after observing $x_i$.


* The soft assignment vector is precisely the conditional expectation $\mathbb{E}[Z_i\mid X_i=x_i]$.


* The M-step updates the cluster parameters using these posterior membership probabilities as weights.



**Conclusion:** Gaussian mixture clustering is probabilistic clustering based on conditional expectations of latent cluster membership variables.

```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

class GMMFinancialSegmenter:
    def __init__(self, n_components=3, random_state=42):
        """
        Initializes the segmenter with a specific number of clusters (components).
        """
        self.n_components = n_components
        self.random_state = random_state
        self.scaler = StandardScaler()
        self.gmm = GaussianMixture(
            n_components=self.n_components,
            random_state=self.random_state,
            covariance_type='full'
        )
        self.X_train_scaled = None
        self.X_test_scaled = None

    def prepare_data(self, df, feature_cols):
        """
        Extracts, standardizes, and splits the dataset into 80% train / 20% test.
        """
        X = df[feature_cols].values

        # 80/20 Train-Test Split
        X_train, X_test = train_test_split(
            X, test_size=0.20, random_state=self.random_state
        )

        # Standardize features using training distribution properties
        self.X_train_scaled = self.scaler.fit_transform(X_train)
        self.X_test_scaled = self.scaler.transform(X_test)

        return self.X_train_scaled, self.X_test_scaled

    def fit(self):
        """
        Fits the GMM model on the training set using the Expectation-Maximization algorithm.
        """
        if self.X_train_scaled is None:
            raise ValueError("Data not prepared. Call prepare_data first.")

        self.gmm.fit(self.X_train_scaled)

        print("--- EM Algorithm Execution Summary ---")
        print(f"Model Converged: {self.gmm.converged_}")
        print(f"Iterations Required: {self.gmm.n_iter_}")

    def evaluate(self):
        """
        Computes the average out-of-sample log-likelihood score on the test data.
        """
        if self.X_test_scaled is None:
            raise ValueError("Data not prepared. Call prepare_data first.")

        avg_log_likelihood = self.gmm.score(self.X_test_scaled)
        print(f"Average Out-of-Sample Log-Likelihood Score: {avg_log_likelihood:.4f}")
        return avg_log_likelihood

    def _generate_contour_grid(self, padding=0.5, resolution=200):
        """
        Helper method to build a 2D meshgrid and compute responsibilities for contour overlays.
        """
        x_min, x_max = self.X_train_scaled[:, 0].min() - padding, self.X_train_scaled[:, 0].max() + padding
        y_min, y_max = self.X_train_scaled[:, 1].min() - padding, self.X_train_scaled[:, 1].max() + padding

        xx, yy = np.meshgrid(np.linspace(x_min, x_max, resolution),
                             np.linspace(y_min, y_max, resolution))
        grid_points = np.c_[xx.ravel(), yy.ravel()]

        # Compute responsibilities: shape (resolution^2, n_components)
        responsibilities = self.gmm.predict_proba(grid_points)
        # Max responsibility for the background soft assignment visualization
        max_resp = np.max(responsibilities, axis=1).reshape(xx.shape)

        return xx, yy, max_resp

    def plot_empirical_density(self, feature_names=None):
        """
        Plot 1: 2D Density Heatmap of raw training data with marginal distributions.
        """
        x_label = feature_names[0] if feature_names else "Feature 1 (Scaled)"
        y_label = feature_names[1] if feature_names else "Feature 2 (Scaled)"

        df_train = pd.DataFrame(self.X_train_scaled, columns=[x_label, y_label])

        fig = px.density_heatmap(
            df_train, x=x_label, y=y_label,
            marginal_x="histogram", marginal_y="histogram",
            title="Figure 1: Empirical 2D Density Heatmap (Training Data)"
            # Removed color_continuous_scale="Viridis" to fix the ValueError
        )
        fig.show()

    def plot_assignments(self, mode="train", feature_names=None):
        """
        Plot 2 & 3: Overlays data points on top of the continuous maximum responsibility contours.
        """
        x_label = feature_names[0] if feature_names else "Feature 1 (Scaled)"
        y_label = feature_names[1] if feature_names else "Feature 2 (Scaled)"

        xx, yy, max_resp = self._generate_contour_grid()
        data_to_plot = self.X_train_scaled if mode == "train" else self.X_test_scaled
        labels = self.gmm.predict(data_to_plot)

        # Base contour map for continuous soft assignment expectations
        fig = go.Figure(data=[
            go.Contour(
                x=xx[0, :], y=yy[:, 0], z=max_resp,
                colorscale="Cividis", contours_coloring="heatmap",
                line_width=0, opacity=0.4,
                colorbar=dict(title="Max Responsibility (γ)")
            )
        ])

        # Overlay discrete predicted assignments as scatter clusters
        for cluster_idx in range(self.n_components):
            cluster_mask = (labels == cluster_idx)
            fig.add_trace(go.Scatter(
                x=data_to_plot[cluster_mask, 0],
                y=data_to_plot[cluster_mask, 1],
                mode='markers',
                name=f'Cluster {cluster_idx + 1}',
                marker=dict(size=6, line=dict(width=0.5, color='white'))
            ))

        title_suffix = "Training Assignment Plot" if mode == "train" else "Out-of-Sample Test Assignment Plot"
        fig.update_layout(
            title=f"Figure {2 if mode=='train' else 3}: {title_suffix}",
            xaxis_title=x_label, yaxis_title=y_label,
            template="plotly_white"
        )
        fig.show()

# --- Verification Simulation ---
if __name__ == "__main__":
    # Generate synthetic multimodal data mimicking financial metrics (Purchases vs. Credit Limit)
    np.random.seed(42)
    c1 = np.random.multivariate_normal([1.5, 2.0], [[0.3, 0.1], [0.1, 0.3]], 400)
    c2 = np.random.multivariate_normal([-1.0, -1.0], [[0.5, -0.1], [-0.1, 0.4]], 350)
    c3 = np.random.multivariate_normal([2.0, -1.5], [[0.4, 0.0], [0.0, 0.5]], 250)
    synthetic_data = np.vstack([c1, c2, c3])

    mock_df = pd.DataFrame(synthetic_data, columns=["PURCHASES", "CREDIT_LIMIT"])

    # Initialize and run pipeline
    segmenter = GMMFinancialSegmenter(n_components=3)
    segmenter.prepare_data(mock_df, feature_cols=["PURCHASES", "CREDIT_LIMIT"])
    segmenter.fit()
    segmenter.evaluate()

    # Generate interactive plots
    segmenter.plot_empirical_density(feature_names=["PURCHASES", "CREDIT_LIMIT"])
    segmenter.plot_assignments(mode="train", feature_names=["PURCHASES", "CREDIT_LIMIT"])
    segmenter.plot_assignments(mode="test", feature_names=["PURCHASES", "CREDIT_LIMIT"])
```
