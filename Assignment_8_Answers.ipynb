# Q. Statistical Mechanics of the Quantum Oscillator

1. The Partition Function

To find the canonical partition function, we sum the Boltzmann factor over all possible quantum states $j$. Given the energy eigenvalues $E_{j}=\hbar\omega(j+1/2)$, the partition function is:$$Z(T)=\sum_{j=0}^{\infty}e^{-\frac{E_{j}}{k_{B}T}}=\sum_{j=0}^{\infty}e^{-\frac{\hbar\omega(j+1/2)}{k_{B}T}} \quad \text{[cite: 7]}$$We can factor out the constant zero-point energy term $e^{-\frac{\hbar\omega}{2k_{B}T}}$ from the summation:$$Z(T)=\frac{e^{-\frac{\hbar\omega}{2k_{B}T}}}{1-e^{-\frac{\hbar\omega}{k_{B}T}}} \quad \text{}$$

Let $x=e^{-\frac{\hbar\omega}{k_{B}T}}$

 Since $x<1$ at any finite temperature, the series converges. The summation is a standard infinite geometric series of the form
 $\sum x^{j}=\frac{1}{1-x}$ ,

  yielding:$$Z(T)=\frac{e^{-\frac{\hbar\omega}{2k_{B}T}}}{1-e^{-\frac{\hbar\omega}{k_{B}T}}} \quad \text{}$$

2. Thermodynamic Potentials

Using the partition function, we can derive the primary thermodynamic potentials.


The Helmholtz Free Energy $F(T)$:$$F(T)=-k_{B}T \ln Z = -k_{B}T \ln\left(\frac{e^{-\frac{\hbar\omega}{2k_{B}T}}}{1-e^{-\frac{\hbar\omega}{k_{B}T}}}\right) \quad \text{}$$$$F(T)=-k_{B}T\left(-\frac{\hbar\omega}{2k_{B}T}-\ln\left(1-e^{-\frac{\hbar\omega}{k_{B}T}}\right)\right) \quad \text{}$$$$F(T)=\frac{\hbar\omega}{2}+k_{B}T \ln\left(1-e^{-\frac{\hbar\omega}{k_{B}T}}\right) \quad \text{}$$The Internal Energy $U(T)$:

 Using $\beta=\frac{1}{k_{B}T}$, the partition function is expressed as $Z=e^{-\beta\frac{\hbar\omega}{2}}(1-e^{-\beta\hbar\omega})^{-1}$.
$$U(T)=-\frac{\partial \ln Z}{\partial\beta}=-\frac{\partial}{\partial\beta}\left(-\beta\frac{\hbar\omega}{2}-\ln\left(1-e^{-\beta\hbar\omega}\right)\right) \quad \text{}$$$$U(T)=\frac{\hbar\omega}{2}+\frac{1}{1-e^{-\beta\hbar\omega}}(\hbar\omega e^{-\beta\hbar\omega}) \quad \text{}$$$$U(T)=\hbar\omega\left(\frac{1}{2}+\frac{1}{e^{\frac{\hbar\omega}{k_{B}T}}-1}\right) \quad \text{}$$

3. Entropy

Entropy is the negative temperature derivative of the Free Energy:$$S(T)=-\left(\frac{\partial F}{\partial T}\right)=-\frac{\partial}{\partial T}\left(\frac{\hbar\omega}{2}+k_{B}T \ln\left(1-e^{-\frac{\hbar\omega}{k_{B}T}}\right)\right) \quad \text{}$$Using the product and chain rules:$$F+TS=\frac{\hbar\omega}{2}+\frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_{B}T}}-1}=U(T) \quad \text{}$$$$S(T)=-k_{B}\ln\left(1-e^{-\frac{\hbar\omega}{k_{B}T}}\right)+\frac{\hbar\omega}{T}\left(\frac{1}{e^{\frac{\hbar\omega}{k_{B}T}}-1}\right) \quad \text{}$$Verification of $U=F+TS$:
$$C_{V}=k_{B}\left(\frac{\hbar\omega}{k_{B}T}\right)^{2}\frac{e^{\frac{\hbar\omega}{k_{B}T}}}{\left(e^{\frac{\hbar\omega}{k_{B}T}}-1\right)^{2}} \quad \text{}$$The $k_{B}T \ln(...)$ terms perfectly cancel out, leaving:$$C_{V}=k_{B}\left(\frac{\hbar\omega}{k_{B}T}\right)^{2}\frac{1}{\left(e^{\frac{\hbar\omega}{2k_{B}T}}-e^{-\frac{\hbar\omega}{2k_{B}T}}\right)^{2}} \quad \text{}$$

4. Heat Capacity and Hyperbolic Form

The constant-volume heat capacity is the temperature derivative of internal energy:$$C_{V}=\left(\frac{\partial U}{\partial T}\right)=\frac{\partial}{\partial T}\left[\hbar\omega\left(\frac{1}{2}+(e^{\frac{\hbar\omega}{k_{B}T}}-1)^{-1}\right)\right] \quad \text{}$$$$C_{V}=\hbar\omega\left[-(e^{\frac{\hbar\omega}{k_{B}T}}-1)^{-2}\left(e^{\frac{\hbar\omega}{k_{B}T}}\cdot\left(-\frac{\hbar\omega}{k_{B}T^{2}}\right)\right)\right] \quad \text{}$$$$C_{V}=k_{B}\left(\frac{\hbar\omega}{k_{B}T}\right)^{2}\frac{e^{\frac{\hbar\omega}{k_{B}T}}}{\left(e^{\frac{\hbar\omega}{k_{B}T}}-1\right)^{2}} \quad \text{}$$Hyperbolic Transformation: Using the identity $\sinh(x)=\frac{e^{x}-e^{-x}}{2}$,

we multiply the numerator and denominator by $e^{-\frac{\hbar\omega}{k_{B}T}}$:$$C_{V}=k_{B}\left(\frac{\hbar\omega}{k_{B}T}\right)^{2}\frac{1}{e^{\frac{\hbar\omega}{k_{B}T}}-2+e^{-\frac{\hbar\omega}{k_{B}T}}} \quad \text{}$$$$C_{V}=k_{B}\left(\frac{\hbar\omega}{k_{B}T}\right)^{2}\frac{1}{\left(e^{\frac{\hbar\omega}{2k_{B}T}}-e^{-\frac{\hbar\omega}{2k_{B}T}}\right)^{2}} \quad \text{}$$$$C_{V}=k_{B}\left(\frac{\hbar\omega}{2k_{B}T}\cdot2\right)^{2}\frac{1}{\left(2 \sinh\left(\frac{\hbar\omega}{2k_{B}T}\right)\right)^{2}} \quad \text{}$$$$C_{V}=k_{B}\left(\frac{\hbar\omega}{2k_{B}T}\right)^{2}\frac{1}{\sinh^{2}\left(\frac{\hbar\omega}{2k_{B}T}\right)} \quad \text{}$$

5. Asymptotic Behavior (High-Temperature Limit)

In the high-temperature limit where $k_{B}T\gg\hbar\omega$, the argument becomes very small.

For $x \rightarrow 0$ where $x=\frac{\hbar\omega}{2k_{B}T}$ , the Taylor expansion gives $\sinh(x)\approx x$.

Substituting this approximation into our hyperbolic heat capacity equation yields:$$C_{V}\approx k_{B}(x)^{2}\frac{1}{x^{2}}=k_{B} \quad \text{}$$Yes, this matches the classical equipartition theorem prediction.

A 1D classical harmonic oscillator has two degrees of freedom (one kinetic, one potential), each contributing $k_{B}/2$ to the heat capacity, yielding a total of $k_{B}$.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Continuous Principle of Maximum Entropy

To find the distribution that maximizes entropy under given constraints, we utilize the continuous Shannon-Jaynes entropy functional with a uniform baseline reference density $m(x)=1$.

The domain is bounded to $X\in(-\infty,\infty)$.Our constraints are:

 3)The Variance: $\mathbb{E}[(X-\mu)^{2}]=\int_{-\infty}^{\infty}p(x)\cdot(x-\mu)^{2}dx=\sigma^{2}$

 We construct the functional Lagrangian $\mathcal{Z}[p]$:$$\mathcal{Z}[p]=-\int_{-\infty}^{\infty}p(x)\ln(p(x))dx-\gamma\left(\int_{-\infty}^{\infty}p(x)dx-1\right)-\lambda_{1}\left(\int_{-\infty}^{\infty}p(x)x \, dx-\mu\right)-\lambda_{2}\left(\int_{-\infty}^{\infty}p(x)(x-\mu)^{2}dx-\sigma^{2}\right)$$Taking the functional Euler-Lagrange derivative with respect to $p(x)$ and setting it to zero:$$\frac{\delta\mathcal{Z}}{\delta p(x)}=-\ln(p(x))-1-\gamma-\lambda_{1}x-\lambda_{2}(x-\mu)^{2}=0 \quad \text{}$$Solving for $p(x)$ gives the maximum entropy exponential form:$$P(Y=1|x)=\frac{1}{1+e^{2.4845x_{1}-1.0421x_{2}-2.3369x_{3}}} \quad \text{}$$To satisfy the condition that the mean $\mathbb{E}[X]=\mu$, the distribution must be symmetric around $\mu$.

  This symmetry forces the linear multiplier to vanish, yielding:λ1​=0

 Evaluating the remaining normalization and variance integrals yields the standard Gaussian parameters:$\lambda_{2}=\frac{1}{2\sigma^{2}}$ $Z=\sqrt{2\pi\sigma^{2}}$

  Substituting these back provides the classic probability density function of the Normal Distribution:$$p^{*}(x)=\frac{1}{\sqrt{2\pi\sigma^{2}}}\exp\left(-\frac{(x-\mu)^{2}}{2\sigma^{2}}\right) \quad \text{}$$

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

Multiclass Softmax Conditional Distribution

Consider the Softmax classifier that uses the standard uniform indicator feature mapping across all choices: $f(x,c)=\mathbb{I}(Y=c)x$.

This yields a multi-class Softmax conditional distribution parameterized by the two weight arrays $w_{0},w_{1}\in\mathbb{R}^{3}$:$$P(Y=c|x)=\frac{\exp(w_{c}\cdot x)}{\sum_{l=0}^{1}\exp(w_{l}\cdot x)}=\frac{e^{w_{c}\cdot x}}{e^{w_{0}\cdot x}+e^{w_{1}\cdot x}} \quad \text{}$$By applying the change of variables $u_{1j}=w_{1j}-w_{0j}$ , we can express the probability for the target Technical destination $(Y=1)$ as:$$P(Y=1|x)=\frac{1}{1+e^{-u_{11}x_{1}-u_{12}x_{2}-u_{13}x_{3}}} \quad \text{}$$Using the relation $\mathbb{E}[x_{j}\mathbb{I}(Y=c)]=\sum_{x\in \mathcal{X}}x_{j}\cdot P(Y=c|x)\cdot P(x)$ , we map the historical constraints across the five scenarios (A, B, C, D, E).

System of Non-Linear Equations: Plugging the macrostate expectations into the system yields:

1.)  $0.05=0.40\left(\frac{e^{u_{11}}}{1+e^{u_{11}}}\right)+0.05\left(\frac{e^{u_{11}+u_{12}}}{1+e^{u_{11}+u_{12}}}\right)$

2.)  $0.15=0.25\left(\frac{e^{u_{12}}}{1+e^{u_{12}}}\right)+0.15\left(\frac{e^{u_{12}+u_{13}}}{1+e^{u_{12}+u_{13}}}\right)+0.05\left(\frac{e^{u_{11}+u_{12}}}{1+e^{u_{11}+u_{12}}}\right)$

3.)  $0.16=0.05\left(\frac{e^{u_{13}}}{1+e^{u_{13}}}\right)+0.15\left(\frac{e^{u_{12}+u_{13}}}{1+e^{u_{12}+u_{13}}}\right)$

Using numerical root finding on these valid bounds yields the exact relative difference parameters:

$u_{11}\approx-2.4845$

$u_{12}\approx1.0421$

$u_{13}\approx2.3369$

Substituting these into the model provides the operational classifier equation:$$p^{*}(S_{T})=\frac{1}{134.72}\cdot\frac{\exp\left(0.1843S_{T}-0.00114S_{T}^{2}\right)}{S_{T}^{0.4215}} \quad \text{}$$

# Q. Asset Pricing and Portfolio Optimization

Maximum Entropy Risk-Neutral Probability Distribution

We maximize the continuous Shannon-Jaynes entropy functional $H_{c}[p]$ subject to the known market constraints. Because asset prices must be positive, our domain is $S_{T}\in(0,\infty)$ and we assume a uniform baseline measure $m(S_{T})=1$.

The Optimization Problem:
$$\max_{p}H_{c}[p]=-\int_{0}^{\infty}p(S_{T})\ln(p(S_{T}))dS_{T} \quad \text{}$$Subject to:

1.)Normalization: $\int_{0}^{\infty}p(S_{T})dS_{T}=1$

2.)Mean: $\int_{0}^{\infty}p(S_{T})\cdot S_{T}dS_{T}=102$

3.)Variance/Energy: $\int_{0}^{\infty}p(S_{T})\cdot S_{T}^{2}dS_{T}=10,600$

4.)Skewness/Crash Risk: $\int_{0}^{\infty}p(S_{T})\cdot \ln(S_{T})dS_{T}=4.60$

Solution Structure: Using Lagrangian calculus, the resulting continuous probability distribution takes the exponential form:$$p^{*}(S_{T})=\frac{1}{Z}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}-\lambda_{3}\ln(S_{T})\right) \quad \text{}$$Using the algebraic identity $\exp(-\lambda_{3}\ln(S_{T}))=S_{T}^{-\lambda_{3}}$ , we can rewrite the asset price distribution:$$p^{*}(S_{T})=\frac{1}{Z}S_{T}^{-\lambda_{3}}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}\right) \quad \text{}$$Where $Z$ is the partition function normalizing constant:$$Z(\lambda_{1},\lambda_{2},\lambda_{3})=\int_{0}^{\infty}S_{T}^{-\lambda_{3}}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}\right)dS_{T} \quad \text{}$$System of Equations and Numerical Solution: The expected values are tied directly to the partial derivatives of the log-partition function:

1.)  $-\frac{\partial \ln Z}{\partial\lambda_{1}}=102\Rightarrow\frac{1}{Z}\int_{0}^{\infty}S_{T}^{1-\lambda_{3}}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}\right)dS_{T}=102 \quad \text{}$

2.)  $-\frac{\partial \ln Z}{\partial\lambda_{2}}=10,600\Rightarrow\frac{1}{Z}\int_{0}^{\infty}S_{T}^{2-\lambda_{3}}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}\right)dS_{T}=10,600 \quad \text{}$

3.)$-\frac{\partial \ln Z}{\partial\lambda_{3}}=4.60\Rightarrow\frac{1}{Z}\int_{0}^{\infty}\ln(S_{T})S_{T}^{-\lambda_{3}}\exp\left(-\lambda_{1}S_{T}-\lambda_{2}S_{T}^{2}\right)dS_{T}=4.60 \quad \text{}$

Minimizing the dual objective function numerically (via BFGS or Newton-Raphson) yields the following parameter values:

$\lambda_{1}\approx-0.1843$ (Shifting the bulk of the distribution forward toward the 102 target)

$\lambda_{2}\approx0.00114$ (Tail-decay preventing infinite expansion)

$\lambda_{3}\approx0.4215$ (Creates a steep left-tail modeling crash fear since it behaves as $\frac{1}{S_{T}^{0.4215}}$)

 $Z\approx134.72$

 The Final Probability Density Curve: Substituting the optimal parameters produces the final unbiased risk-neutral distribution:$$p^{*}(S_{T})=\frac{1}{134.72}\cdot\frac{\exp\left(0.1843S_{T}-0.00114S_{T}^{2}\right)}{S_{T}^{0.4215}} \quad \text{}$$

```python
import numpy as np
import plotly.graph_objects as go

# 1. Define the solved MaxEnt parameters
lambda_1 = -0.1843
lambda_2 = 0.00114
lambda_3 = 0.4215
Z = 134.72
expected_value = 102.0

# 2. Generate asset price range (avoiding absolute zero to prevent log/division errors)
S_T = np.linspace(0.1, 200, 1000)

# 3. Compute the MaxEnt Probability Density Function (PDF)
pdf = (1.0 / Z) * (S_T ** (-lambda_3)) * np.exp(-lambda_1 * S_T - lambda_2 * (S_T**2))

# 4. Create the interactive Plotly figure
fig = go.Figure()

# Add the main PDF curve
fig.add_trace(
    go.Scatter(
        x=S_T,
        y=pdf,
        mode='lines',
        name='MaxEnt Asset PDF',
        line=dict(color='#636EFA', width=3),
        hovertemplate='Asset Price: $%{x:.2f}<br>Probability Density: %{y:.5f}<extra></extra>'
    )
)

# Add a vertical dashed line for the Expected Price (Market Forward)
fig.add_trace(
    go.Scatter(
        x=[expected_value, expected_value],
        y=[0, max(pdf) * 1.05],
        mode='lines',
        name=f'Expected Value (${expected_value:.1f})',
        line=dict(color='#EF553B', width=2, dash='dash'),
        hovertemplate='Expected Forward Price: $102.00<extra></extra>'
    )
)

# 5. Update layout aesthetics
fig.update_layout(
    title=dict(
        text='<b>Maximum Entropy Risk-Neutral Asset Price Distribution</b><br><sup>Asymmetric crash risk fear modeled via log-space constraints</sup>',
        font=dict(size=16)
    ),
    xaxis=dict(
        title='Asset Price at Maturity $S_T$ ($)',
        gridcolor='rgba(200, 200, 200, 0.2)',
        zeroline=False
    ),
    yaxis=dict(
        title='Probability Density $p^*(S_T)$',
        gridcolor='rgba(200, 200, 200, 0.2)',
        zeroline=False
    ),
    template='plotly_white',
    width=900,
    height=550,
    legend=dict(
        yanchor="top",
        y=0.95,
        xanchor="right",
        x=0.95,
        bgcolor="rgba(255, 255, 255, 0.6)"
    )
)

# Show the plot inside the notebook cell
fig.show()
```
