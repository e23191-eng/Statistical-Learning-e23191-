#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

### Q1.3

By definition, $E[X\mid Y]$ must be $\sigma(Y)$-measurable. This means it can depend only on the value of $Y$, not on the full outcome $\omega$. That is $E[X\mid Y]=g(Y)$ for some measurable map $g$.

###Q1.4

To find $a$ and $b$, we use the defining property of conditional expectation.

On the set $\{\omega_1,\omega_2\}$,

$$
\int_{\{\omega_1,\omega_2\}} E[X\mid Y]\,dP=\int_{\{\omega_1,\omega_2\}} X\,dP.
$$

The left-hand side is

$$
a\cdot P(\{\omega_1,\omega_2\})=a(0.50+0.10)=0.60a.

The right-hand side is

$$
0\cdot 0.50 + 1\cdot 0.10 = 0.10.
$$

So

$$
0.60a=0.10,
$$

hence

$$
a=\frac{0.10}{0.60}=\frac{1}{6}.
$$

On the set $\{\omega_3,\omega_4\}$,

$$
\int_{\{\omega_3,\omega_4\}} E[X\mid Y]\,dP=\int_{\{\omega_3,\omega_4\}} X\,dP.

The left-hand side is

$$
b\cdot P(\{\omega_3,\omega_4\})=b(0.20+0.20)=0.40b.

0\cdot 0.20 + 1\cdot 0.20 = 0.20.

$$
0.40b=0.20,

hence

$$
b=\frac{0.20}{0.40}=\frac{1}{2}.

Therefore

$$

E[X\mid Y]=

###Q1.6

Using the product rule $p_{X|Y}(x \mid y) = \frac{P(X=x, Y=y)}{P(Y=y)}$, we calculate the conditional probability of fraud ($X=1$) for each risk class:

* **For $Y=0$:**
  $$p_{X|Y}(1 \mid 0) = \frac{P(X=1, Y=0)}{P(Y=0)} = \frac{P(\{\omega_2\})}{0.60} = \frac{0.10}{0.60} = \mathbf{1/6}$$
* **For $Y=1$:**
  $$p_{X|Y}(1 \mid 1) = \frac{P(X=1, Y=1)}{P(Y=1)} = \frac{P(\{\omega_4\})}{0.40} = \frac{0.20}{0.40} = \mathbf{1/2}$$


So once the bank knows the risk class, the fraud probability changes:

  $$

  $$

* for high-risk customers,
  $$
  P(\text{fraud}\mid Y=1)=\frac{1}{2}=0.50.
  $$


This is exactly the kind of quantity a fraud detection system would use for risk scoring.

###Q1.7

Now apply your discrete summary formula:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

* **Evaluation at $y=0$:**
  $$\mathbb{E}[X\mid Y=0] = (0 \cdot p_{X|Y}(0 \mid 0)) + (1 \cdot p_{X|Y}(1 \mid 0)) = 0 + (1 \cdot 1/6) = \mathbf{1/6}$$
* **Evaluation at $y=1$:**
  $$\mathbb{E}[X\mid Y=1] = (0 \cdot p_{X|Y}(0 \mid 1)) + (1 \cdot p_{X|Y}(1 \mid 1)) = 0 + (1 \cdot 1/2) = \mathbf{1/2}$$

###Q1.8

We compute the Minimum Mean Square Error (MMSE), which represents the optimal error when using the estimate $\mathbb{E}[X|Y]$. As derived previously, this is given by the expected conditional variance:
$$\text{MSE}_{\text{min}} = \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \mathbb{E}[\mathrm{Var}(X \mid Y)].$$

First, we find the conditional variance for each risk class $y \in \{0, 1\}$. Since $X$ is a Bernoulli indicator variable given $Y=y$, the variance is $p(1-p)$:

For $Y=0$: $\mathrm{Var}(X \mid Y=0) = \frac{1}{6} \left(1 - \frac{1}{6}\right) = \frac{1}{6} \cdot \frac{5}{6} = \frac{5}{36}$.

For $Y=1$: $\mathrm{Var}(X \mid Y=1) = \frac{1}{2} \left(1 - \frac{1}{2}\right) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}.$

Now, we take the expectation over the distribution of $Y$:

\begin{align*}\mathbb{E}[\mathrm{Var}(X \mid Y)] &= \mathrm{Var}(X \mid Y=0)P(Y=0) + \mathrm{Var}(X \mid Y=1)P(Y=1) \\
&= \left(\frac{5}{36} \cdot 0.60\right) + \left(\frac{1}{4} \cdot 0.40\right) \\
&= \left(\frac{5}{36} \cdot \frac{3}{5}\right) + \left(\frac{1}{4} \cdot \frac{2}{5}\right) \\
&= \frac{1}{12} + \frac{1}{10} = \frac{5+6}{60} = \mathbf{\frac{11}{60} \approx 0.1833}.
\end{align*}
This value, $0.1833$, is the lowest possible average squared error the bank can achieve. Any other estimate based on risk class would result in an error strictly greater than this.

###Q1.9

If the bank ignores the risk class $Y$ and uses the global mean $\mathbb{E}[X] = 0.30$ as its estimate, the resulting error is the **Total Variance of $X$**.

**Calculating the Total MSE:**

When the estimate is a constant $c = \mathbb{E}[X]$, the Mean Square Error is simply the variance of $X$:
$$\text{MSE}_{\text{global}} = \mathbb{E}[(X - \mathbb{E}[X])^2] = \mathrm{Var}(X)$$

Since $X$ is a Bernoulli random variable with $P(X=1) = 0.30$:
$$\mathrm{Var}(X) = p(1-p) = 0.30(0.70) = \mathbf{0.21}$$

**Comparing the Errors:**

* **Optimal Error (with $Y$):** $\text{MSE}_{\text{min}} = \mathbb{E}[\mathrm{Var}(X \mid Y)] = \frac{11}{60} \approx \mathbf{0.1833}$
* **Global Error (without $Y$):** $\mathrm{Var}(X) = \frac{12.6}{60} = \mathbf{0.21}$

By using the information in $Y$, the bank reduces its prediction error from **0.21** to **0.1833**.

**Relationship to the Law of Total Variance:**

The **Law of Total Variance** decomposes the total uncertainty into two distinct parts:
$$\mathrm{Var}(X) = \mathbb{E}[\mathrm{Var}(X \mid Y)] + \mathrm{Var}(\mathbb{E}[X \mid Y])$$

In the context of this example, the terms represent:

*  **$\mathbb{E}[\mathrm{Var}(X \mid Y)] = 0.1833$ (Unexplained Variance):** This is the "Optimal Error" you calculated in Point 8. It represents the noise that remains even after we use the best possible model.
    
* **$\mathrm{Var}(\mathbb{E}[X \mid Y])$ (Explained Variance):**
This represents the "Information Gain" or the amount of error we **removed** by using $Y$.
Let's calculate it:
* $\mathbb{E}[X \mid Y]$ is $1/6$ with probability $0.6$ and $1/2$ with probability $0.4$.
* $\mathrm{Var}(\mathbb{E}[X \mid Y]) = \mathbb{E}[(\mathbb{E}[X \mid Y])^2] - (\mathbb{E}[X])^2$
* $= [(\frac{1}{36} \cdot 0.6) + (\frac{1}{4} \cdot 0.4)] - (0.3)^2 = [0.0166 + 0.1] - 0.09 = \mathbf{0.0267}$

>The Final Breakdown:
$$0.21 (\text{Total Error}) = 0.1833 (\text{Remaining Error}) + 0.0267 (\text{Error Reduction})$$


### Q1.10 Quantifying Information Gain

To calculate these information-theoretic quantities, we first need to summarize the joint distribution $P(X, Y)$ and the conditional probabilities.

1. The Joint Distribution $P(X, Y)$

Based on the mappings $X(\omega)$ and $Y(\omega)$, we can group the outcomes:
* $P(X=0, Y=0) = P(\{\omega_1\}) = 0.50$
* $P(X=1, Y=0) = P(\{\omega_2\}) = 0.10$
* $P(X=0, Y=1) = P(\{\omega_3\}) = 0.20$
* $P(X=1, Y=1) = P(\{\omega_4\}) = 0.20$

The marginal distributions are:
* **For $Y$:** $P(Y=0) = 0.60$, $P(Y=1) = 0.40$
* **For $X$:** $P(X=0) = 0.70$, $P(X=1) = 0.30$

---

2. Calculate Conditional Entropy $H(X \mid Y)$

Conditional entropy is the weighted average of the entropy of $X$ within each group of $Y$.

**Step A: Entropy of $X$ given $Y=0$**
$$P(X=1 \mid Y=0) = \frac{1}{6}, \quad P(X=0 \mid Y=0) = \frac{5}{6}$$
$$H(X \mid Y=0) = -\left[ \frac{1}{6} \log_2 \frac{1}{6} + \frac{5}{6} \log_2 \frac{5}{6} \right] \approx 0.650 \text{ bits}$$

**Step B: Entropy of $X$ given $Y=1$**
$$P(X=1 \mid Y=1) = \frac{1}{2}, \quad P(X=0 \mid Y=1) = \frac{1}{2}$$
$$H(X \mid Y=1) = -\left[ \frac{1}{2} \log_2 \frac{1}{2} + \frac{1}{2} \log_2 \frac{1}{2} \right] = 1.00 \text{ bit}$$

**Step C: Final Conditional Entropy**
$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$
$$H(X \mid Y) = (0.60 \times 0.650) + (0.40 \times 1.00) = 0.390 + 0.40 = \mathbf{0.790 \text{ bits}}$$

---

3. Calculate Mutual Information $I(X; Y)$

Mutual Information is the reduction in uncertainty: $I(X; Y) = H(X) - H(X \mid Y)$.

**Step A: Calculate $H(X)$**
$$H(X) = -(0.70 \log_2 0.70 + 0.30 \log_2 0.30) \approx \mathbf{0.881 \text{ bits}}$$

**Step B: Subtract $H(X \mid Y)$**
$$I(X; Y) = 0.881 - 0.790 = \mathbf{0.091 \text{ bits}}$$

---

4. Interpretation:

* **$H(X) = 0.881$:** This is the total uncertainty the bank has about fraud before looking at the risk class.
* **$H(X \mid Y) = 0.790$:** This is the uncertainty remaining after the bank knows if a customer is "Low-Risk" or "High-Risk."
* **$I(X; Y) = 0.091$:** This represents the "Information Gain." By knowing the customer's risk class, the bank has reduced its uncertainty about fraud by approximately **10.3%** ($0.091 / 0.881$).

> **In Practical Terms:** Knowing the risk class $Y$ provides a small but measurable amount of information about the fraud indicator $X$. The fact that $I(X; Y) > 0$ proves that $Y$ is a useful feature for the bank's fraud detection model.

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

##  Answer

### Conditional Density of the Multivariate Gaussian Derivation

To find the conditional density $f_{X|Y}(x \mid y) = \frac{f_{X,Y}(x, y)}{f_Y(y)}$, we partition the joint covariance matrix and its inverse using the **Schur Complement**.

#### 1. Matrix Inversion & Identity
The joint covariance matrix is partitioned as:
$$\Sigma = \begin{bmatrix} \Sigma_{XX} & \Sigma_{XY} \\ \Sigma_{YX} & \Sigma_{YY} \end{bmatrix}$$

Using the Schur complement of $\Sigma_{YY}$, defined as $\Sigma_{X|Y} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}$, we can factor the inverse matrix $\Sigma^{-1}$ as:
$$\Sigma^{-1} = \begin{bmatrix} I & -\Sigma_{XY}\Sigma_{YY}^{-1} \\ 0 & I \end{bmatrix}^T \begin{bmatrix} \Sigma_{X|Y}^{-1} & 0 \\ 0 & \Sigma_{YY}^{-1} \end{bmatrix} \begin{bmatrix} I & -\Sigma_{XY}\Sigma_{YY}^{-1} \\ 0 & I \end{bmatrix}$$

#### 2. Decomposition of the Joint Quadratic Form
Let $\tilde{x} = x - \mu_X$ and $\tilde{y} = y - \mu_Y$. The quadratic form in the joint density exponent expands to:
$$(z - \mu)^T \Sigma^{-1} (z - \mu) = \begin{bmatrix} \tilde{x} \\ \tilde{y} \end{bmatrix}^T \Sigma^{-1} \begin{bmatrix} \tilde{x} \\ \tilde{y} \end{bmatrix}$$

Substituting our factored matrix, the expression splits into two separate quadratic terms:
$$(z - \mu)^T \Sigma^{-1} (z - \mu) = (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*) + (y - \mu_Y)^T \Sigma_{YY}^{-1} (y - \mu_Y)$$

where the conditional mean vector is:
$$\mu^* = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$

#### 3. Determinant Factorization
The determinant of the block matrix evaluates to:
$$|\Sigma| = |\Sigma_{YY}| \cdot |\Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}| = |\Sigma_{YY}| \cdot |\Sigma_{X|Y}|$$

#### 4. Final Substitution
Dividing the joint density $f_{X,Y}(x,y)$ by the marginal density $f_Y(y)$ cancels out the terms related exclusively to $Y$:

$$f_{X|Y}(x \mid y) = \frac{\frac{1}{(2\pi)^{n/2}|\Sigma_{X|Y}|^{1/2}}\exp\left(-\frac{1}{2}(x-\mu^*)^T\Sigma_{X|Y}^{-1}(x-\mu^*)\right) \cdot \frac{1}{(2\pi)^{m/2}|\Sigma_{YY}|^{1/2}}\exp\left(-\frac{1}{2}(y-\mu_Y)^T\Sigma_{YY}^{-1}(y-\mu_Y)\right)}{\frac{1}{(2\pi)^{m/2}|\Sigma_{YY}|^{1/2}}\exp\left(-\frac{1}{2}(y-\mu_Y)^T\Sigma_{YY}^{-1}(y-\mu_Y)\right)}$$

$$f_{X|Y}(x \mid y) = \frac{1}{(2\pi)^{n/2} |\Sigma_{X|Y}|^{1/2}} \exp\left(-\frac{1}{2} (x - \mu^*)^T \Sigma_{X|Y}^{-1} (x - \mu^*)\right)$$

Thus, the conditional distribution is also Gaussian:
$$X \mid Y=y \sim \mathcal{N}(\mu^*, \Sigma_{X|Y})$$

# Q3) Continuous Fault Monitoring of a Spindle Motor

**Answer the followings:**

Compute the conditional expectation $E[X \mid Y]$.

**5. Information Theory: Differential Entropy**

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



## Answer

**1. Joint Distribution of $(X, Y)$:**

Since $X$ and $Q$ are independent normal random variables, their sum $Y = X + Q$ is also normally distributed.
* **Mean:** $E[Y] = E[X] + E[Q] = 0 + 0 = 0$.
* **Variance:** Since $X \perp Q$, $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$.
* **Covariance:** $\text{Cov}(X, Y) = E[XY] - E[X]E[Y] = E[X(X+Q)] = E[X^2] + E[XQ]$.
    * Since $E[X^2] = \text{Var}(X) = 1$ and $E[XQ] = E[X]E[Q] = 0$:
    * **$\text{Cov}(X, Y) = 1$**.

The joint distribution is:
$$\begin{pmatrix} X \\ Y \end{pmatrix} \sim \mathscr{N} \left( \begin{pmatrix} 0 \\ 0 \end{pmatrix}, \begin{pmatrix} 1 & 1 \\ 1 & 5 \end{pmatrix} \right)$$

---

**2. Information Geometry:**

$X$ is not $\sigma(Y)$-measurable because there is no deterministic function $g$ such that $X = g(Y)$. For any observed $Y=y$, there are infinitely many pairs $(x, q)$ such that $x+q=y$.
**Practically:** A high power reading ($Y$) is ambiguous; it could be a "Thermal Fault" (high $X$) or simply a "Hard Material" (high $Q$). The sensor $Y$ cannot distinguish between these two physical states.

---

**3. Conditional Expectation (The Best Estimate):**

For bivariate normal variables, the best mean-square estimate is:
$$E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0)$$
$$\mathbf{E[X \mid Y] = 0.2 Y}$$
**Interpretation:** Because the noise ($Q$) is 4 times more volatile than the signal ($X$), we only "trust" 20% of the power fluctuation as being caused by temperature.

---

**4. Verification of the Residual:**

Let $\widehat{X} = 0.2 Y$. The residual is $Z = X - 0.2 Y$.
* **Mean:** $E[Z] = E[X] - 0.2 E[Y] = 0 - 0 = 0$.
* **Orthogonality:** $\text{Cov}(Z, Y) = E[ZY] = E[(X - 0.2Y)Y] = E[XY] - 0.2 E[Y^2]$.
    * $E[XY] = 1$ and $E[Y^2] = 5$.
    * $\text{Cov}(Z, Y) = 1 - 0.2(5) = 1 - 1 = 0$.
The error is uncorrelated with our observation, meaning we have extracted all possible information from $Y$.

---

**5. Differential Entropy:**

Recall $h(\mathscr{N}(\mu, \sigma^2)) = \frac{1}{2}\ln(2\pi e \sigma^2)$.
* **Marginal $h(X)$:** With $\sigma_X^2 = 1$:
$$h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$
* **Conditional $h(X|Y)$:** The conditional variance is
$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1}{5} = 0.8.$$
$$h(X|Y) = \frac{1}{2}\ln(2\pi e \cdot 0.8) \approx 1.3073 \text{ nats}$$

---

**6. Mutual Information:**

$$I(X; Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(1) - \frac{1}{2}\ln(0.8) = \frac{1}{2}\ln\left(\frac{1}{0.8}\right) = \frac{1}{2}\ln(1.25) \approx \mathbf{0.1116 \text{ nats}}$$
**Percentage Uncertainty Removed:** This is often measured by the reduction in variance. We went from a variance of $1.0$ to $0.8$ (a **20% reduction**). If the noise $Q$ increased to $\sigma_Q^2 = 100$, $I(X;Y)$ would approach 0, as the temperature signal would be "drowned out."

---

**7. Diagnostic Interpretation:**

The manager's suggestion to use $X = Y$ is a **Naive Estimator**.
* If we use $X = Y$, our error variance would be $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, our error variance is only $0.8$.

**Shrinkage:** The factor $0.2$ is called a shrinkage estimator. It tells us that when a sensor is noisy, the "best" guess is to be conservative and pull the reading back toward the historical average (zero). If the power spikes, we don't immediately assume the tool is melting; we assume most of that spike is just the material being tough.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.stats import multivariate_normal, norm

# --- 1. Parameters for the Spindle Problem ---
# X: Temperature ~ N(0, 1)
# Q: Torque ~ N(0, 4)
# Y: Power = X + Q => Y ~ N(0, 5)
mu_x, mu_y = 0, 0
var_x, var_q = 1, 4
var_y = var_x + var_q
cov_xy = var_x  # Cov(X, X+Q) = Var(X) = 1
rho = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))

# Shrinkage Factor (Gain)
k = cov_xy / var_y  # 1/5 = 0.2

# --- 2. Generate Data for Visualization ---
# Sample points to show the "Cloud" of data
np.random.seed(42)
n_samples = 1000
x_samples = np.random.normal(mu_x, np.sqrt(var_x), n_samples)
q_samples = np.random.normal(0, np.sqrt(var_q), n_samples)
y_samples = x_samples + q_samples
x_hat = k * y_samples  # E[X|Y]
residual = x_samples - x_hat

# Grids for Contours
x_grid = np.linspace(-4, 4, 200)
y_grid = np.linspace(-8, 8, 200)
X_m, Y_m = np.meshgrid(x_grid, y_grid)
pos = np.dstack((X_m, Y_m))
Z = multivariate_normal([mu_x, mu_y], [[var_x, cov_xy], [cov_xy, var_y]]).pdf(pos)

# --- 3. Create Simulation Plot ---
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Joint Distribution: X vs Y<br><sup>The Line of Best Estimate (Shrinkage)</sup>",
        "Estimation Error (Residual) vs Y<br><sup>Visualizing the Orthogonality Principle</sup>"
    )
)

# --- LEFT PLOT: Joint X, Y and the Estimator ---
# Density Contour
fig.add_trace(go.Contour(
    x=x_grid, y=y_grid, z=Z, colorscale='Blues', showscale=False, opacity=0.6,
    hoverinfo='skip'
), row=1, col=1)

# Scatter of actual realizations
fig.add_trace(go.Scatter(
    x=x_samples, y=y_samples, mode='markers',
    marker=dict(color='rgba(0,0,0,0.1)', size=3),
    name="Realized (x, y)"
), row=1, col=1)

# The Best Estimate Line E[X|Y] = 0.2Y
y_line = np.linspace(-8, 8, 10)
fig.add_trace(go.Scatter(
    x=k * y_line, y=y_line,
    line=dict(color='Red', width=4),
    name="E[X|Y] = 0.2Y (Optimal)"
), row=1, col=1)

# The Naive Estimator Line X = Y
fig.add_trace(go.Scatter(
    x=y_line, y=y_line,
    line=dict(color='Black', width=2, dash='dash'),
    name="X = Y (Naive)"
), row=1, col=1)

# --- RIGHT PLOT: Orthogonality (Residual vs Y) ---
fig.add_trace(go.Scatter(
    x=y_samples, y=residual, mode='markers',
    marker=dict(color='Purple', opacity=0.4, size=4),
    name="Residual (X - 0.2Y)"
), row=1, col=2)

# Horizontal line at 0 to show no correlation
fig.add_trace(go.Scatter(
    x=y_line, y=np.zeros_like(y_line),
    line=dict(color='Black', width=2),
    name="Zero Mean Error"
), row=1, col=2)

# --- Layout and Styling ---
fig.update_layout(
    height=600,
    title_text="Spindle Motor Diagnostic Simulation: Extracting Signal from Noise",
    template="plotly_white",
    hovermode='x'
)

fig.update_xaxes(title_text="True Temperature (X)", row=1, col=1)
fig.update_yaxes(title_text="Observed Power (Y)", row=1, col=1)
fig.update_xaxes(title_text="Observed Power (Y)", row=1, col=2)
fig.update_yaxes(title_text="Residual Error (Z)", row=1, col=2)

fig.show()
```
